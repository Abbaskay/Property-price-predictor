```python
dummies = pd.get_dummies(df2.Locality)
dummies.head(4)
```

```python
df4 = pd.concat([df3,dummies],axis='columns')
df4.head(3)
```

```python
df5 = df4.drop('Locality',axis='columns')
df5.head(2)
```

```python
df5.shape
```

```python
df5 = df5.replace(np.nan, 0)
```

```python
X = df5.drop('Price',axis='columns')
X.head()
```

```python
Y = df5.Price
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=10)
```

```python
from sklearn. linear_model import LinearRegression

clf = LinearRegression()

```

```python
X_train
```

```python
df5
```

```python
Y_train = df5['Price'].iloc[:1007]
```

```python
Y_train
```

```python
clf.fit(X_train,Y_train)

clf.score (X_test,Y_test)
```

```python
from sklearn.model_selection import ShuffleSplit 
from sklearn.model_selection import cross_val_score
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
cross_val_score(LinearRegression(), X, Y, cv=cv)
```

```python
def predict_price(Locality,Area,Bathroom,BHK,Parking):    
    loc_index = np.where(X.columns==Locality)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = Area
    x[1] = Bathroom
    x[2] = BHK
    x[3] = Parking
    
    if loc_index >= 0:
        x[loc_index] = 1

    return clf.predict([x])[0]
```
